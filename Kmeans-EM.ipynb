```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
#for zero mean and unit covariance 
from sklearn.preprocessing import StandardScaler
data=pd.read_csv("income.csv")
```

```python
features=["Age","Income($)"]
sub_data=data[features]
data_1=sub_data.to_numpy()
#for zero mean and unit covariance 
scaler = StandardScaler()
data_scaled = scaler.fit_transform(sub_data)
```

```python
kmeans=KMeans(n_clusters=3, init="k-means++" , random_state=0).fit(data_scaled)
centroids = kmeans.cluster_centers_
labels=kmeans.labels_
```

```python
plt.scatter(data_1[:,0], data_1[:,1], c=labels ,cmap='viridis')
#plt.scatter(centroids[:,0], centroids[:,1], marker="*",s=50, c='black')
plt.title('KMeans')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

# Load the data
data = pd.read_csv("income.csv")
features = ["Age", "Income($)"]
sub_data = data[features]
sub_data_1=sub_data.to_numpy()

# Scale the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(sub_data_1)

# Apply EM clustering with K=3
em = GaussianMixture(n_components=3, covariance_type='full', random_state=0)
em.fit(data_scaled)

# Get the predicted cluster labels and cluster centers
labels = em.predict(data_scaled)
centers = em.means_

# Visualize the results
plt.scatter(data_scaled[:,0], data_scaled[:,1], c=labels, cmap='viridis')
plt.scatter(centers[:,0], centers[:,1], marker='*', s=300, c='black')
plt.title('EM Clustering with PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
